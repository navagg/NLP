##### Finding Bigrams: To understand it better, Lets run it on first three words 

```python
for w in words [:3]:
    for ch1, ch2 in zip (w[:], w[1:]):
        print(ch1, ch2)
```

###### But we must have start and end character, Lets modified the code accordingly

```python
for w in words [:1]:
    # we are adding special character at start and end of each word
    ws=["<S>"] + list(w) + ["<E>"]
    for ch1, ch2 in zip (ws[:], ws[1:]):
        print(ch1, ch2)
```

#### Now we are able to generate bigrams, So we have to do it for full set of words. Rather than printing , we can add them into a dictionary

```python
b={}
for w in words [:4]:
    # we are adding special character at start and end of each word
    ws=["<S>"] + list(w) + ["<E>"]
    for ch1, ch2 in zip (ws[:], ws[1:]):
        # Make a Tuple of selected Ch1 and ch2
        bigram=(ch1,ch2)
        # Add into the dictionary, b.get gives count if bigram is already in dictionary otherwise zero
        b[bigram]=b.get(bigram,0)+1
print(b)
```

##### Now Lets Run the code for all the words

```python
b={}
for w in words [:]:
    ws=["<S>"] + list(w) + ["<E>"]
    for ch1, ch2 in zip (ws[:], ws[1:]):
        bigram=(ch1,ch2)
        b[bigram]=b.get(bigram,0)+1
```

#### Now to deal with this lets use one token for start and end. Let it be '.' Further, we will take this as first token and give numbers to alphabets a to z from 1 to 26.

```python
chars= list('.abcdefghijklmnopqrstuvwxyz')
stoi={s:i for i,s in enumerate(chars)}
stoi

# Lets also create itos, we will use later
itos={i:s for i,s in enumerate (chars)}
itos
```

```python
N=torch.zeros((27,27), dtype=torch.int32)
for w in words [:]:
    ws=  ws=["."] + list(w) + ["."]
    for ch1, ch2 in zip (ws[:], ws[1:]):
        ix1=stoi[ch1]
        ix2=stoi[ch2]
        N[ix1,ix2]+=1
```

```python
print(N[:5,:5])
```

#### Now Lets create  a probability distribution, Divide each term by sum of all terms

```python
p=N[0].float()
p=p/p.sum()
p
```

##### Now lets understand, what is torch Generator: It just generae Pseudo random numbers based on a seed value

```python
g=torch.Generator().manual_seed(160014)
p=torch.rand(3, generator=g)
#p
p=p/p.sum()
p
```

#### Now This can be taken as a the Probability Distribution. Based on this Probability distribution, if we generate numbers, 10% will be '0', 2% will be 1 and 87% will be 2.
Lets check it

```python
torch.multinomial(p, num_samples=100, replacement= True, generator=g)
```

#### With this understanding, Now we want to extract digits from our Array N with their probability of occurance. Lets write code for that

```python

g=torch.Generator().manual_seed(2147583637)
# Tosave result
out=[]
ix=0
while True:
    p=N[ix].float()
    p=p/p.sum()
    ix=torch.multinomial(p, num_samples=1, replacement= True, generator=g).item()
    out.append(itos[ix])
    if ix == 0:
        break
print(''.join(out))

# You can change the Generator seed to get different result
```

#### Lets create 10 such names
#### And Lets just optimize it bit by using the Broadcasting

```python
# This will run faster, we will also add 1 to avoid Zero Probaility
P=(N+1).float()
P/=P.sum(1, keepdims=True)
```

```python
for i in range(10):
    out=[]
    ix=0
    while True:
        p=P[ix]
        ix=torch.multinomial(p, num_samples=1, replacement= True, generator=g).item()
        out.append(itos[ix])
        if ix == 0:
            break
    print(''.join(out))
    
```

### So Above Model is just predicting next letter based on Probability districution defined using Generator

#### Now Lets find the log likelihood of next character after each character
#### We will first calculate Probability, We will take its log.
#### We will then use negative log likelhood.

```python
# Use the same code and Modify it
log_likelihood=0.0
count=0
for w in words [:3]:
    ws=  ws=["."] + list(w) + ["."]
    for ch1, ch2 in zip (ws[:], ws[1:]):
        ix1=stoi[ch1]
        ix2=stoi[ch2]
        prob=P[ix1,ix2]
        logprob=torch.log(prob)
        log_likelihood+=logprob
        count+=1
        print(f'{ch1}{ch2}: {prob:4f} {logprob:4f}')
print(f'{log_likelihood}')
# Negative Log Likelihood
nll=-log_likelihood
print(f'{nll}')
# Average negative log likelihood
print(f'{nll/count}')
```
